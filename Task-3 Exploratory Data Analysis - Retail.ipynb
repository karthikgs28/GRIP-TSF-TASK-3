### Effect of Ship Mode and Segment on Profit

```python
fig = plt.figure(figsize=(10,7))
ship = df.groupby('Ship Mode').mean().reset_index()
plt.bar(ship['Ship Mode'],ship['Profit'],width = 0.3)
plt.title('Profit Made in each Ship Mode')
plt.show()
ship
```

```python
fig = plt.figure(figsize=(10,7))
seg = df.groupby('Segment').mean().reset_index()
plt.bar(seg['Segment'],seg['Profit'],width = 0.3)
plt.title('Profit Made in each Segment')
plt.show()
seg
```

##### we can say that Ship Mode has minimal effect on the profit, and in terms of Segment H.O. has the high profti margin followed by Corporate and Consumer.

### Let's Look At How Each Region Performs on the basis of Profit

```python
fig = plt.figure(figsize=(10,8))
region = df.groupby('Region').mean().reset_index()
plt.pie(region['Profit'],autopct='%1.2f%%',explode = (0.04, 0.04, 0.04,0.04),colors = ('blue','green','red','brown'))
plt.legend(region['Region'])
plt.title('Profit Made in Each Sector')
plt.show()
region
```

```python
df.groupby(['Region','Category']).mean()
```

### Lets See the top 10 States in terms of Profit

```python
fig = plt.figure(figsize=(8,5))
state = df.groupby('State').mean().reset_index().sort_values(by = 'Profit',ascending=False).head(10)
plt.bar(state['State'],state['Profit'],width = 0.3,color = 'blue')
plt.xticks(rotation = 45)
plt.title('Profit Made in Each State')
plt.show()
state
```

### Let's Look At How Each Sector Performs on the basis of Profit

```python
fig = plt.figure(figsize=(10,8))
cate = df.groupby('Category').mean().reset_index()
plt.pie(cate['Profit'],autopct='%1.2f%%',explode = (0.08, 0.04, 0.06),colors = ('blue','green','red'))
plt.legend(cate['Category'])
plt.title('Profit Made in Each Sector')
plt.show()
cate
```

```python
fig = plt.figure(figsize=(10,8))
cate = df[df['Profit']<0].groupby('Category').mean().reset_index()
plt.pie(cate['Profit']*-1,autopct='%1.2f%%',explode = (0.04, 0.05, 0.03),colors = ('blue','green','red'))
plt.legend(cate['Category'])
plt.title('Loses Made in Each Sector')
plt.show()
cate
```

```python
#top 10 profits
df.sort_values(by = 'Profit').tail(10)
```

### Let us take a broad look at what subcategories contribute to what amount of profit -

### Here's a look on how much much each sub-category is discounted -

```python
fig = plt.figure(figsize=(8,5))
sub = df.groupby(['Sub-Category']).mean().reset_index().sort_values(by = 'Discount',ascending=False)
plt.bar(sub['Sub-Category'],sub['Discount'],width = 0.3,color = 'red')
plt.xticks(rotation = 45)
plt.title('')
plt.show()
sub
```

### Furniture

```python
figure, axis = plt.subplots(1, 2,figsize=(15,6))
cat = df.groupby(['Category','Sub-Category']).mean().reset_index()
sub_c = cat[cat['Category'] == 'Furniture']
axis[0].bar(sub_c['Sub-Category'],sub_c['Profit'],width=0.2)
axis[0].set_title('Profit Made in Each Sub-Category in Furniture')
axis[0].axhline(0, color='black')
axis[1].bar(sub_c['Sub-Category'],sub_c['Sales'],width=0.2,color = 'brown')
axis[1].set_title("Sales Made in Each Sub-Category in Furniture")
plt.show()
sub_c.iloc[:,1:]
```

### Technology

```python
figure, axis = plt.subplots(1, 2,figsize=(15,6))
cat = df.groupby(['Category','Sub-Category']).mean().reset_index()
sub_c = cat[cat['Category'] == 'Technology']
axis[0].bar(sub_c['Sub-Category'],sub_c['Profit'],width=0.2,color = 'red')
axis[0].set_title('Profit Made in Each Sub-Category in Technology')
axis[1].bar(sub_c['Sub-Category'],sub_c['Sales'],width=0.2,color = 'green')
axis[1].set_title("Sales Made in Each Sub-Category in Technology")
plt.show()
sub_c.iloc[:,1:]
```

### Office Supplies

```python
figure, axis = plt.subplots(1, 2,figsize=(15,6))
cat = df.groupby(['Category','Sub-Category']).mean().reset_index()
sub_c = cat[cat['Category'] == 'Office Supplies']
axis[0].bar(sub_c['Sub-Category'],sub_c['Profit'],width=0.2)
axis[0].set_title('Profit Made in Each Sub-Category in Office Supplies')
axis[0].set_xticklabels(labels = sub_c['Sub-Category'],rotation = 30)
axis[0].axhline(0, color='black')
axis[1].bar(sub_c['Sub-Category'],sub_c['Sales'],width=0.2,color = 'orange')
axis[1].set_xticklabels(sub_c['Sub-Category'],rotation = 30)
axis[1].set_title("Sales Made in Each Sub-Category in Office Supplies")
plt.show()
sub_c.iloc[:,1:]
```
